```python
# Dropping unnecessary columns
df = dfraw.drop(columns=[
    'Location', 'Total Land Area (acres)', 'Number of Cows', 
    'Farm Size','Product ID', 'Quantity (liters/kg)',
    'Price per Unit', 'Total Value', 'Shelf Life (days)',
    'Storage Condition', 'Production Date', 
    'Expiration Date','Quantity Sold (liters/kg)',
    'Quantity in Stock (liters/kg)', 
    'Minimum Stock Threshold (liters/kg)',
    'Reorder Quantity (liters/kg)'])

# Display dataframe after dropping
df.head()
```

```python
# Change column names for easier calling

## Replace all columns into lowercase and spaces into underscore "_"
df.columns = df.columns.str.lower().str.replace(' ', '_')

## Rename some of the column names
df.rename(columns={
    'product_name': 'product',
    'price_per_unit_(sold)': 'unit_price_sold',
    'approx._total_revenue(inr)': 'total_revenue'
}, inplace=True)
df.columns
```

## **Exploratory Data Analysis (EDA)**

Pada tahap ini akan menganalisis masing-masing kolom serta memberikan analisa visual dan deskriptif sesuai pertanyaan-pertanyaan yang akan tertera pada masing-masing kolom.

### **Date / Waktu Penjualan**

Mengidentifikasi pola musiman atau tren tahunan: *Apakah ada tahun atau bulan tertentu di mana penjualan cenderung meningkat atau menurun?*

Untuk itu, dapat melihat distribusi penjualan berdasarkan tanggal atau *Date* dengan membuat visualisasi penjualan per tahun dan bulan untuk melihat tren musiman penjualan dari waktu ke waktu.

```python
# Convert date column to date time data type
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
```

#### **Annual Sales**

```python
# Annual sales analysis

## Creating a new column for year and grouping
df['year'] = df['date'].dt.year
yearly_sales = df.groupby('year')['total_revenue'].sum()

## Visualization
sns.set_theme(style="darkgrid", palette="viridis")

plt.figure(figsize=(12, 7))
sns.barplot(x=yearly_sales.index, y=yearly_sales.values, palette="viridis", width=0.6)
plt.title('Tren Penjualan Tahunan', fontsize=16, color='white', weight='bold')
plt.xlabel('Tahun', fontsize=14, color='white')
plt.ylabel('Total Pendapatan (INR)', fontsize=14, color='white')
plt.xticks(color='white', fontsize=12)
plt.yticks(color='white', fontsize=12)
plt.grid(True, color='#444444', axis='y', linestyle='--', linewidth=0.7)
plt.figtext(0.5, -0.05, "Catatan: Nominal Pendapatan Berupa 10.000.000 INR", ha="right", color="white", fontsize=10)

## Adding numbers on top of each bars
for index, value in enumerate(yearly_sales.values):
    plt.text(index, value + 0.05 * value, f'{int(value):,}', ha='center', color='yellow', fontsize=11, fontweight='bold')

## Visualization background customization
plt.gca().set_facecolor('#222222')
plt.gcf().set_facecolor('#111111')
plt.show()
```

Berdasarkan visualisasi, data diambil dari tahun 2019 sampai tahun 2022. Penjualan susu dapat terlihat menurun dari tahun ke tahun, ini mungkin disebabkan oleh beberapa faktor:
* **Kenaikan Harga:** Biaya produksi yang tinggi akibat kenaikan harga pakan, energi, dan transportasi mendorong kenaikan harga susu, yang pada akhirnya mengurangi daya beli konsumen.
* **Gangguan Rantai Pasok:** Konflik geopolitik dan masalah logistik mengganggu pasokan bahan baku dan distribusi produk susu. (*)
* **Perubahan Preferensi Konsumen:** Meningkatnya kesadaran akan kesehatan, tren diet, dan persaingan dari produk substitusi mengubah pola konsumsi masyarakat.
* **Penyakit Ternak:** Wabah penyakit pada ternak mengurangi produksi susu dan mendorong kenaikan harga. (*)

Penurunan penjualan susu di India merupakan hasil dari interaksi kompleks antara berbagai faktor ekonomi, sosial, dan lingkungan. Untuk mengatasi masalah ini, diperlukan upaya bersama dari pemerintah, produsen susu, dan konsumen untuk mencari solusi jangka panjang, seperti diversifikasi produksi, pengembangan produk inovatif, dan peningkatan efisiensi produksi.

(*) *Reference:* [CNBC](https://www.cnbcindonesia.com/news/20230525150409-4-440610/kabar-buruk-dari-india-susunya-mau-meledak)

#### **Monthly Sales**

Penjualan bulanan yang akan divisualisasikan adalah penjualan bulanan pada tahun 2022 dikarenakan data yang lebih baru sehingga lebih relevan.

```python
# Monthly sales analysis (2022)

## Creating a new column for month and grouping
df['month'] = df['date'].dt.month
monthly_sales_2022 = df[df['year'] == 2022].groupby('month')['total_revenue'].sum()

## Visualization
plt.figure(figsize=(12, 7))
sns.lineplot(x=monthly_sales_2022.index, y=monthly_sales_2022.values, color=sns.color_palette("viridis", as_cmap=True)(0.7), linewidth=3, marker='o', markersize=8)
plt.title('Tren Penjualan Bulanan di Tahun 2022', fontsize=16, color='white', weight='bold')
plt.xlabel('Bulan', fontsize=14, color='white')
plt.ylabel('Total Pendapatan (INR)', fontsize=14, color='white')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], color='white', fontsize=12)
plt.yticks(color='white', fontsize=12)
plt.grid(True, color='#444444', linestyle='--', linewidth=0.7)
plt.figtext(0.5, -0.05, "Catatan: Nominal Pendapatan Berupa 1.000.000 INR", ha="right", color="white", fontsize=10)
## Adding value on top of each marker/month
for index, value in enumerate(monthly_sales_2022.values):
    plt.text(index + 1, value + 0.05 * value, f'{int(value):,}', ha='center', color='yellow', fontsize=11, fontweight='bold')

plt.gca().set_facecolor('#222222')
plt.gcf().set_facecolor('#111111')
plt.show()
```

Grafik penjualan produk susu di India tahun 2022 menunjukkan fluktuasi yang cukup signifikan. Penjualan mencapai puncaknya pada awal tahun dan kemudian mengalami penurunan yang cukup drastis di pertengahan tahun. Meskipun terjadi penurunan, penjualan produk susu menunjukkan tanda-tanda pemulihan pada akhir tahun.

### **Product / Produk Terlaris**

Mengetahui produk susu mana yang paling laris di pasaran: *Produk mana yang paling sering dibeli pelanggan?*

Melihat distribusi penjualan berdasarkan produk (`product`) yang paling banyak terjual berdasarkan pendapatan yang dihasilkan mulai dari tahun 2019 - 2022.

```python
# Products sold analysis

## Grouping top product by revenue
top_products = df.groupby('product')['total_revenue'].sum().sort_values(ascending=False)

## Visualization
plt.figure(figsize=(12, 8))
sns.barplot(y=top_products.index, x=top_products.values, palette="viridis")
plt.title('Produk Susu Terlaris Berdasarkan Total Pendapatan', fontsize=16, color='white', weight='bold')
plt.xlabel('Total Pendapatan (INR)', fontsize=14, color='white')
plt.ylabel('Nama Produk', fontsize=14, color='white')
plt.xticks(color='white', fontsize=12)
plt.yticks(color='white', fontsize=12)
plt.grid(True, color='#444444', axis='x', linestyle='--', linewidth=0.7)
plt.figtext(0.5, -0.05, "Catatan: Nominal Pendapatan Berupa 1.000.000 INR", ha="right", color="white", fontsize=10)

## Adding value on each bar
for index, value in enumerate(top_products.values):
    plt.text(value + 0.05 * value, index, f'{int(value):,}', ha='left', color='yellow', fontsize=11, fontweight='bold')

plt.gca().set_facecolor('#222222')
plt.gcf().set_facecolor('#111111')
plt.show()
```

Berdasarkan *bar plot* diatas, dapat disimpulkan bahwa pasar produk susu di India sangat besar dan beragam. *Curd*, *Butter*, dan *Lassi* merupakan produk susu yang paling populer dan memiliki *Market Share* yang signifikan. Faktor budaya, ketersediaan, harga, dan manfaat kesehtatan menjadi kunci keberhasilan ketiga produk ini.

### **Customer Location / Lokasi Kustomer**

Lokasi kustomer dengan penjualan tertinggi: *Kota mana saja yang memiliki penjualan tertinggi?*

Memberikan visualisasi dan menjelaskan lebih dalam lagi mengapa kota-kota dengan peringkat teratas memiliki penjualan yang tinggi.

Terdapat perbedaan yang signifikan antara dua kota (Chandigarh dan Delhi) dengan pendapatan tertinggi dengan kota lainnya. Karena itu, hanya kedua kota tersebut yang akan dibahas mengapa memiliki penjualan yang tinggi. Menampilkan visualisasi saluran penjualan berdasarkan penjualan produk, serta memberikan implikasi dan rekomendasi terhadap saluran penjualan mengingat perusahaan *WicaksanaFoods* merupakan perusahaan yang baru, sehingga saluran penjualan atau distribusi penjualan merupakan suatu fondasi pertama dalam berjualan.

Grafik di atas memberikan gambaran yang jelas mengenai bagaimana produk susu dipasarkan di India dan saluran mana yang paling efektif dalam menghasilkan pendapatan.     - **Rekomendasi :**

* **Pertumbuhan Penjualan Online**
    - **Implikasi :**
